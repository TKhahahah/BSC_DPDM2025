# ระบุข้อมูลสูญหายแต่ละสถานี แสดงจำนวนและสัดส่วน

```python
relevant_columns = ['rainfall_datetime', 'rainfall_datetime_calc', 'rainfall_value', 'tele_station_id', 'measure_datetime', 'tele_station_oldcode', 'station_code', 'value']

missing_data_by_column_and_station = pd.DataFrame()

for col in relevant_columns:
    missing_values = combined_data_filtered_cleaned.groupby('station_id')[col].apply(lambda x: x.isnull().sum())
    total_entries = combined_data_filtered_cleaned.groupby('station_id').size()

    summary = pd.DataFrame({
        'missing_count': missing_values,
        'total_entries': total_entries,
        'missing_proportion': missing_values / total_entries
    })
    summary = summary[summary['missing_count'] > 0]
    if not summary.empty:
        summary['column'] = col
        missing_data_by_column_and_station = pd.concat([missing_data_by_column_and_station, summary])

if not missing_data_by_column_and_station.empty:
    missing_data_by_column_and_station = missing_data_by_column_and_station.sort_values(by=['column', 'missing_proportion'], ascending=[True, False])
    print("Missing data per station per column (count and proportion):")
    print(missing_data_by_column_and_station)
else:
    print("No missing data found in the specified columns for any station.")
```

#คัดสถานีที่มีข้อมูลสูญหาย >= 20% ออก แสดงรายชื่อสถานีที่ถูกคัดออกและจำนวนทีีเหลือ

```python
stations_to_remove_overall = missing_data_by_column_and_station[missing_data_by_column_and_station['missing_proportion'] >= 0.2].index.unique().tolist()

combined_data_filtered_cleaned_final = combined_data_filtered_cleaned[~combined_data_filtered_cleaned['station_id'].isin(stations_to_remove_overall)]

print(f"Stations removed due to >= 20% missing data in any column: {stations_to_remove_overall}")
print(f"Number of stations removed: {len(stations_to_remove_overall)}")
print(f"Original shape of combined_data_filtered_cleaned: {combined_data_filtered_cleaned.shape}")
print(f"New shape of combined_data_filtered_cleaned_final: {combined_data_filtered_cleaned_final.shape}")
```

#เติมข้อมูลที่หายไปด้วยค่ากลาง(Median) ตาม Day of Year จาก daily_rain แยกตามสถานี แล้วตรวจสอบว่าไม่มี rocords หายไปอีก

```python
# Acknowledge the empty DataFrame issue from the previous step
print("Note: The 'combined_data_filtered_cleaned_final' DataFrame is empty after the previous filtering step that removed all stations.")
print("Proceeding with missing data imputation on 'combined_data_filtered' instead.")

# 1. Prepare daily_rain for median calculation
daily_rain_imputation = daily_rain.copy()
daily_rain_imputation['measure_datetime'] = pd.to_datetime(daily_rain_imputation['measure_datetime'], format='mixed', errors='coerce')
daily_rain_imputation['day_of_year'] = daily_rain_imputation['measure_datetime'].dt.dayofyear

# Rename columns in daily_rain to match combined_data_filtered for consistent merging
daily_rain_imputation = daily_rain_imputation.rename(columns={'station_code': 'station_id', 'data': 'rainfall_value'})

# Calculate median rainfall_value from daily_rain for each station_id and day_of_year
median_rainfall_by_day_station = daily_rain_imputation.groupby(['station_id', 'day_of_year'])['rainfall_value'].median().reset_index()
median_rainfall_by_day_station = median_rainfall_by_day_station.rename(columns={'rainfall_value': 'median_historical_rainfall'})

# 2. Prepare combined_data_filtered for imputation
combined_data_for_imputation = combined_data_filtered.copy()
combined_data_for_imputation['measure_datetime'] = pd.to_datetime(combined_data_for_imputation['measure_datetime'], format='mixed', errors='coerce')
combined_data_for_imputation['day_of_year'] = combined_data_for_imputation['measure_datetime'].dt.dayofyear

# 3. Merge and Impute
# Merge combined_data_for_imputation with historical medians
combined_data_for_imputation = pd.merge(
    combined_data_for_imputation,
    median_rainfall_by_day_station,
    on=['station_id', 'day_of_year'],
    how='left'
)

# Fill missing rainfall_value with the merged median_historical_rainfall
# It's important to only fill NaNs that exist. If median_historical_rainfall is NaN for a given station/day, it won't fill.
combined_data_for_imputation['rainfall_value'] = combined_data_for_imputation['rainfall_value'].fillna(
    combined_data_for_imputation['median_historical_rainfall']
)

# Drop the temporary 'day_of_year' and 'median_historical_rainfall' columns as they are no longer needed
combined_data_for_imputation = combined_data_for_imputation.drop(columns=['day_of_year', 'median_historical_rainfall'])

# Update the main combined_data_filtered to reflect the imputed values
combined_data_filtered = combined_data_for_imputation.copy()

# 4. Verification
print(f"\nNumber of missing 'rainfall_value' records after imputation: {combined_data_filtered['rainfall_value'].isnull().sum()}")
```

เราใช้ median แทน mean เพราะ
ข้อมูลฝนมีค่าผิดปกติ (ฝนตกหนักบางวัน)
การกระจายข้อมูลมักเบ้ ไม่เป็นปกติ
